# Web Scrapping


```python
from bs4 import BeautifulSoup
import requests
```

```python
url=BeautifulSoup('https://www.nrb.org.np/forex/','html.parser')
soup=requests.get(url)
code=BeautifulSoup(soup.text,'lxml')
```

```python
soup  #NO ERROR
```

```python
# Making it more beautiful
codes=(code.prettify())
```

```python
tags=code.find_all('table',attrs={'class':'table table-hover table-borderless'})
tags
```

```python
data=[]
for i in tags:
  data.append(i.text.split('\n'))

data
```

```python
data=data[1:]

```

```python
final_data=[]
for i in data:
  final_data=i[4:]

final_data
```

```python
lst=[]
for i in range(len(final_data)):
  if final_data[i]!='':
    lst.append(final_data[i])

lst

```

```python
final_list=[]
for i in range(len(lst)):
  if (i+1)%4==0:
    final_list.append(lst[i-3:i+1])



```

```python
final_list
```

```python
import csv
try:
  file=open('usddata.csv','w')
  x=csv.writer(file)
  x.writerows(final_list)
  file.close()
except:
  print('File already exists')
```
